```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid coordinates
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Define vector components for converging flow pattern
U = -X
V = Y

# Create quiver plot
plt.figure()
plt.title('Quiver Plot with Converging Flow Pattern')
plt.quiver(X, Y, U, V)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```
